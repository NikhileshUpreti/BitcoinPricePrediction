```python
df=df[['btc_market_price','btc_market_cap','btc_n_transactions','btc_miners_revenue','btc_cost_per_transaction','btc_difficulty','btc_hash_rate','btc_cost_per_transaction_percent']]
```

```python
mean_btc_difficulty=df['btc_difficulty'].mean()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('btc_market_price',axis=1)
```

```python
x
```

```python
y=df['btc_market_price']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
predictor = LinearRegression(n_jobs=-1)
```

```python
predictor.fit(x_train,y_train)
```

```python
outcome=predictor.predict(x_test)
```

```python
coeff= predictor.coef_
```

```python
print("Outcome:",outcome)
print("Coefficient:",coeff)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
y_pred=outcome
```

```python
t=round(mean_squared_error(y_test,y_pred))
```
